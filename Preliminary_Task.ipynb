```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
data=pd.read_csv('/gdrive/My Drive/musk_csv.csv')
```

# Preprocessing 

```
pd.DataFrame(data).head()
```

```
data=data.drop(['molecule_name','conformation_name'],axis=1) #dropping 1st two columns
```

```
data.head()
```

```
data.isna().sum() #checking for null values
```

```
import pandas_profiling as pp
#pp.ProfileReport(data)
```

```
sns.heatmap(data.corr(),annot=True)  #plotting heatmap
```

```
#dropping features having high correlation
corr_matrix = data.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
# Drop features 
data=data.drop(data[to_drop], axis=1)
data.head()
```

# Splitting Data

```
X=data.iloc[:,1:-1]
Y=data.iloc[:,-1]
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,random_state=0,test_size=0.2)
```

# Scaling

```
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
X_train=sc_x.fit_transform(X_train)
X_test=sc_x.fit_transform(X_test)
```

# Using Classification algorithms:

1. Support Vector Classifier- This gave accuracy of  94.72
2. Logistic Regression- This gave accuracy of 93.86

```
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

classifier=SVC(kernel='linear',random_state=0)
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred)
```

```
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred)
```

# Using Dense Layers

```
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten,Activation

```

```
model=Sequential()
model.add(Dense(64,input_shape=(X_train.shape[1],)))  

model.add(Activation('relu')) #specifying activation function

model.add(Dense(1)) 
model.add(Activation('sigmoid')) #acti of o/p layer

model.summary()
model.compile(optimizer='adam',loss='binary_crossentropy',
              metrics=['accuracy'])
history=model.fit(X_train,y_train,batch_size=16,
          nb_epoch=25,
          verbose=2,
          validation_data=(X_test,y_test)
         
         )
```

```
score=model.evaluate(X_test,y_test,verbose=0)
print(score)
```

```
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show();
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show();
```

```
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import f1_score, precision_score, recall_score
print("f1_score:",f1_score(y_test,model.predict_classes(X_test),))
print("recall:",recall_score(y_test,model.predict_classes(X_test),))
print("Test Loss:",score[0])
print("Test Accuracy:",score[1])
```
